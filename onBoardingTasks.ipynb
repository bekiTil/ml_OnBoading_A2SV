**This section is about pandas for extracting data and using matplotlib for visualizing the data we extracted from the csv**

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataframe = pd.read_csv("imdb_top_1000.csv")
```

```python
dataframe
```

# 1.  **What are the column names?**

```python
columns_names = dataframe.columns.tolist()
```

```python
columns_names
```

# 2. **What is the average rating of the movies?**

First Find the Rating column

```python
rating_column = dataframe['IMDB_Rating']

```

```python

rating_mean = rating_column.mean()
```

```python
rating_mean
```

# 3. **How many unique genres are there?**

finding the genre column

```python
genre_column = dataframe['Genre']
```

```python
unique_genres = genre_column.unique()
```

```python
unique_genres.tolist()
```

The length of the unique_genres is going to be the number of unique geners are there in the dataset

```python
unique_genres_count =len(unique_genres)
```

# **Saving my results**

```python
results_df = pd.DataFrame({'Column Names': columns_names,
                           'Average Rating': rating_mean,
                           'Unique Genres Count': unique_genres_count})
```

```python
results_df
```

**Save my results as a csv file.**

```python
results_df.to_csv('results.csv', index=False)
```

# **Plotting**

# Plot a histogram of the genre distribution

```python

dataframe['Genre'].value_counts().plot(kind='bar', figsize=(10, 6))
plt.title('Genre Distribution of Movies')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

**How many movies from each genre are there in this list**

```python
genre_counts = dataframe['Genre'].value_counts()
```

```python
genre_counts
```

# **Plot a scatterplot for number of movies vs. year**

```python
movies_per_year = dataframe['Released_Year'].value_counts().sort_index()

movies_per_year_df = pd.DataFrame({'Year': movies_per_year.index, 'Number of Movies': movies_per_year.values})

movies_per_year_df.plot(kind='scatter', x='Year', y='Number of Movies', figsize=(10, 6))
plt.title('Number of Movies vs. Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```
